## 1- Quais São as Categorias de Filmes Mais Comuns no IMDB?

Quais são os principais tipos (categorias) dos títulos (filmes)?

```python
consulta1 = "SELECT type, COUNT(*) AS COUNT FROM titles GROUP BY type"
print(consulta1)

resultado1_1 = pd.read_sql_query(consulta1, conn)
display(resultado1_1)

resultado1_1['percentual'] = (resultado1_1['COUNT'] / resultado1_1['COUNT'].sum()) * 100
resultado1 = resultado1_1
```

```python
resultado1_1.sort_values(by= 'percentual', ascending=False)
```

```python
dict_others ={}

procura1 = resultado1_1[resultado1_1['percentual'] > 5]

dict_others.update({'COUNT': resultado1_1[resultado1_1['percentual'] < 5]['COUNT'].sum()}) #jeito 1 de fazer
dict_others['percentual'] = resultado1[resultado1['percentual'] < 5]['percentual'].sum()   #jeito2 de fazer
dict_others.update({'type': 'others'})
```

```python
procura1 = procura1.append(dict_others,ignore_index=True)
procura1 = procura1.sort_values(by='percentual',ascending=False)
procura1
```

```python
label = [f"{procura1['type'].iloc[i]} [{round(procura1['percentual'].iloc[i],2)}%]" for i in range(0,len(procura1))]

print(label)
```

## 2- Qual o Número de Títulos Por Gênero?

Vamos calcular o número de filmes por gênero e entregar o resultado em valor percentual.

```python
consulta2= "SELECT genres FROM titles WHERE type='movie' GROUP BY genres "
resultado2= pd.read_sql_query(consulta2, conn)
resultado2
```

```python
resultado2['genres'] = resultado2['genres'].str.lower()
resultado2
```

```python
list_unique_genre = []
for line in range(0,len(resultado2['genres'])):
    new_list = resultado2['genres'].iloc[line].split(',')
    for term in range(0,len(new_list)):
        term_list = new_list[term]
        if term_list not in list_unique_genre:
            list_unique_genre.append(term_list)
list_unique_genre
```

```python
padrao = '\\b[\\w-]+\\b' 
# \   indique que caracteres especiais entram 
# []  conjnto de caracteres
# -   (Se usar [-a] ou [a-]) indica q '-' deve ser considerado como palavras unica e não dentro de uma só e tbm não será sepador
# \w  indca que o proximo caracter deve ser incluso dentro da palavra então 
# [\\w-] = considera caracter especiais e será o '-' para ser em uma única palavra (exemplo-exemplo)
# \b  string vazia no inicio - é a fronteira quando usa \w
# +   indica q a proxima deve entrar (Ex: ab+ corresponde 'a' e os proximos 'b's q são diferente de zero)

extrator = CountVectorizer(token_pattern = padrao, lowercase=True, analyzer = 'word')
extrator = extrator.fit(resultado2['genres'])

get_feature_names = extrator.get_feature_names()
all_genre = extrator.transform(resultado2['genres'])

df_generos = pd.DataFrame(all_genre.todense(), columns = get_feature_names, index = resultado2['genres'].index)
# display(df_generos)
print(df_generos.sum())
```

```python
unique_genres = df_generos.sum().drop(index='n')
unique_genres = pd.DataFrame(unique_genres, columns = ['count'])
unique_genres.sort_values('count',ascending=False,inplace=True)

unique_genres['percentual'] = round((unique_genres['count']/df_generos.shape[0])*100,4)
unique_genres
```

```python
sns.set(rc={'figure.figsize':(13,11)})
sns.barplot(x = unique_genres['percentual'], y=unique_genres.index,palette="plasma")

for i in range(0,len(unique_genres.index)):
    
    plt.text(3, i+0.12, f"Quant. Filmes = {unique_genres['count'][i]}" , fontsize = 8, weight='semibold',color = 'white',
         bbox = dict(facecolor = 'black', alpha = 0.5, ))

plt.annotate('Highest', xy = (18, 0.5), 
             fontsize = 14, xytext = (18.5, 2), 
             arrowprops = dict(facecolor = 'black'),
             color = 'blue')
plt.ylabel('Gênero')             
plt.xlabel("\nPercentual de Filmes (%)")
plt.title('\nNúmero (Percentual) de Títulos Por Gênero\n')
```

## 3- Qual a Mediana de Avaliação dos Filmes Por Gênero?

Mediana de avaliação dos filmes.

```python
consulta3 ="SELECT ratings.title_id, titles.genres,  rating \
            FROM ratings  \
            JOIN titles    \
            ON ratings.title_id = titles.title_id \
            WHERE premiered <= 2022 AND type = 'movie' "
df3 = pd.read_sql_query(consulta3, conn)
df3.head()
```

```python
def retorna_generos(df3):
    temp = df3['genres'].dropna()
    vetor = CountVectorizer(token_pattern = '(?u)\\b[\\w-]+\\b', analyzer = 'word').fit(temp)
    generos_unicos =  vetor.get_feature_names()
    generos_unicos = [genre for genre in generos_unicos if len(genre) > 1]
    return generos_unicos
```

```python
generos_unicos = retorna_generos(df3)
generos_unicos
```

```python
genero_counts = []
genero_ratings = []

for item in generos_unicos:
    consulta = 'SELECT COUNT(rating) FROM ratings JOIN titles ON ratings.title_id=titles.title_id WHERE genres LIKE ' + '\'' + '%' +item+ '%' + '\' AND type=\'movie\''
    resultado = pd.read_sql_query(consulta, conn) 
    genero_counts.append(resultado.values[0][0])

  
     # Retorna a avaliação de filmes por gênero
    consulta = 'SELECT rating, genres FROM ratings JOIN titles ON ratings.title_id=titles.title_id WHERE genres LIKE ' + '\'' + '%' +item+ '%' + '\' AND type=\'movie\' '
    resultado = pd.read_sql_query(consulta, conn)
    genero_ratings.append(np.median(resultado['rating'])) #aqui ele ja calcula a média com base nos generos separados pois já tinha feito na query
    
```

```python
df_median_rating_genre = pd.DataFrame({'genres':generos_unicos,
                                       'count_rating': genero_counts,
                                       'median_rating': genero_ratings,
                                       'count_films' :unique_genres.sort_index()['count'].values},
                                        index = np.arange(1, 29))

df_median_rating_genre = df_median_rating_genre.sort_values(by= 'median_rating', ascending=False)
df_median_rating_genre.reset_index(drop=True,inplace=True)
df_median_rating_genre
```

```python
median_ranting = df_median_rating_genre['median_rating']
median_ranting = median_ranting.reset_index(drop=True)
median_ranting
```

```python
sns.set(rc={'figure.figsize':(13,11)})
# plt.figure(figsize = (16,8))
sns.barplot(x = df_median_rating_genre['median_rating'].sort_values(ascending=False), y=df_median_rating_genre['genres'],palette="winter")

for i in range(0,len(median_ranting)):
    
    plt.text(1, i +0.12, f"Quant. Filmes = {unique_genres['count'][i]}" , fontsize = 8, weight='semibold',color = 'white')
    plt.text(3, i+0.12, f"Quant. Avaliações = {df_median_rating_genre['count_rating'][i]}" , fontsize = 8, weight='semibold',color = 'white')
    plt.text(median_ranting[i], i+0.12, round(median_ranting[i],2), fontsize = 10 ,color = 'black')
    
plt.ylabel('Gênero')             
plt.xlabel("\nRanting de Filmes (%)")
plt.title('\nRating de Títulos Por Gênero\n')
```

## 4- Qual a Mediana de Avaliação dos Filmes Em Relação ao Ano de Estréia?

Mediana de avaliação dos filmes.

```python
consulta4 ="SELECT titles.premiered, rating, votes \
            FROM ratings  \
            JOIN titles    \
            ON ratings.title_id = titles.title_id \
            WHERE premiered <= 2022 AND type = 'movie' \
            ORDER BY premiered, rating"

df4 = pd.read_sql_query(consulta4, conn)
df4.head(10)
```

```python
anos = list(df4['premiered'].unique())

anos_ratings = [np.median(df4[df4['premiered'] == year]['rating']) for year in anos]
anos_votes = [float(df4[df4['premiered'] == year]['votes'].count()) for year in anos]
```

```python
df_rating_year = pd.DataFrame({'year':anos,
                              'median_rating': anos_ratings,
                              'count_votes' : anos_votes }, index = np.arange(126))
df_rating_year
```

```python
sns.set(rc={'figure.figsize':(13,11)})
sns.boxplot(x=df_rating_year['year'], y = df_rating_year['median_rating'],palette="winter", orient='h')    
plt.ylabel('Gênero')             
plt.xlabel("\nRanting de Filmes (%)")
plt.title('\nRating de Títulos Por Gênero\n')
```

```python
plt.figure(figsize = (16,8))
plt.plot(df_rating_year['year'], df_rating_year['median_rating'])
plt.xlabel('\nAno')
plt.ylabel('Mediana de Avaliação')
plt.title('\nMediana de Avaliação dos Filmes Em Relação ao Ano de Estréia\n')
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.plot(df_rating_year['year'], df_rating_year['count_votes']) 
plt.xlabel('\nAno')
plt.ylabel('Quantidade de votos')
plt.title('\nQuantidade de Avaliações dos Filmes Em Relação ao Ano de Estréia\n')
```


<span style='background:lightblue' color='withe'>**ESCALONANDO TUDO E COMPARANDO A RELAÇÃO DE FILMES + VOTADOS E FILMES + LANÇADOS EM RELAÇÃO AO GÊNERO**</span>

- A relação em quantidade de filmes e quantidade de votos não faz muito sentido, foi feito apenas para treinar as consultas, plot e escolha de um gênero pelo usuário

```python
def seleciona_genero(genero):
    for item in genero:
        consulta = 'SELECT COUNT(votes) as votes, premiered YEAR FROM titles JOIN ratings ON ratings.title_id = titles.title_id WHERE genres LIKE '+ '\''+'%'+item+'%'+'\' AND type=\'movie\' AND Year <=2022 GROUP BY Year'
        resultado = pd.read_sql_query(consulta, conn)
        min_max_scaler = MinMaxScaler((0,1))
        df_scalonada = min_max_scaler.fit_transform(resultado)
        plt.plot(resultado['YEAR'], df_scalonada[:,0],'red')

    for item in genero:
        consulta = 'SELECT COUNT(*) as count, premiered YEAR FROM titles  WHERE genres LIKE '+ '\''+'%'+item+'%'+'\' AND type=\'movie\' AND Year <=2022 GROUP BY Year'
        resultado = pd.read_sql_query(consulta, conn)
        min_max_scaler = MinMaxScaler((0,1))
        df_scalonada = min_max_scaler.fit_transform(resultado)
        plt.plot(resultado['YEAR'], df_scalonada[:,0],'m')


    plt.legend(labels = ['Quat. Votos','Quant. Filmes'])
    plt.xlabel('\nAno')
    plt.ylabel('Dados escalonados com base na quantidade')
    plt.title('\n Proporção de quantidade de votos dos usuários por quantidade de filmes do gênero: '+ genero[0].upper())
    plt.show()
```

## 7- Qual a Relação Entre Duração e Gênero?

Relação entre duração do filme e gênero.

```python
consulta7 = "SELECT original_title, premiered, runtime_minutes as Runtime, genres from titles WHERE type = 'movie' AND Runtime != 'NaN' ORDER BY Runtime DESC "
resultado7 = pd.read_sql_query(consulta7, conn)
resultado7.head()
```

```python
genero_unico3 = retorna_generos(resultado7)
```

```python
genero_runtime = []
for item in genero_unico3:
        consulta = 'SELECT (runtime_minutes) as Runtime FROM titles WHERE genres LIKE '+ '\''+'%'+item+'%'+'\' AND type=\'movie\'AND Runtime!=\'NaN\' '
        resultado = pd.read_sql_query(consulta, conn)
        genero_runtime.append(np.median(resultado['Runtime']))
```

```python
new_df = pd.DataFrame(genero_runtime, columns = ['median_runtime'])
new_df['genre'] = genero_unico3
new_df.sort_values(by='median_runtime', inplace=True, ascending=False)
new_df
```

```python
sns.barplot(y = new_df['genre'],x =new_df['median_runtime'])

for i in range(0,len(new_df.index)):
    
#     plt.text(1, i +0.12, f"Quant. Filmes = {unique_genres['count'][i]}" , fontsize = 8, weight='semibold',color = 'white')
#     plt.text(3, i+0.12, f"Quant. Avaliações = {df_median_rating_genre['count_rating'][i]}" , fontsize = 8, weight='semibold',color = 'white')
#     plt.text(median_ranting[i], i+0.12, round(median_ranting[i],2), fontsize = 10 ,color = 'black')
    
    plt.text(new_df["median_runtime"].sort_values(ascending=False).iloc[i],i + 0.25,round(new_df["median_runtime"][new_df.index[i]], ))

```

## 8- Qual o Número de Filmes Produzidos Por País?

Número de filmes produzidos por país.

```python
consulta8 = '''
            SELECT region, COUNT(*) Number_of_movies FROM 
            akas JOIN titles ON 
            akas.title_id = titles.title_id
            WHERE region != 'None'
            AND type = \'movie\'
            GROUP BY region
            ''' 
resultado8 = pd.read_sql_query(consulta8, conn)
resultado8.sort_values('Number_of_movies', ascending=False,inplace=True)
```

```python
resultado8.head()
```

```python
list(pycountry.countries)
```

```python
dic = {}            
for i in range(0,len(resultado8)):
    country = pycountry.countries.get(alpha_2=resultado8['region'].iloc[i])   
    if pycountry.countries.get(alpha_2=resultado8['region'].iloc[i]):
        dic[resultado8['region'].iloc[i]] = [country.name, resultado8['Number_of_movies'].iloc[i]]
        
new_df = pd.DataFrame.from_dict(dic,orient='index', columns=['country','Number_of_movies'])
new_df
```

```python
plt.figure(figsize = (16,8))
plt.plot(new_df['Number_of_movies'][:20],new_df['country'][:20])
sns.barplot(y=new_df['country'][:20], x=new_df['Number_of_movies'][:20])
plt.ylabel('País')             
plt.xlabel('\nNúmero de Filmes')
plt.title('\nNúmero de Filmes Produzidos Por País\n')
```

```python
nomes_paises = []
contagem = []
for i in range(resultado8.shape[0]):
    try:
        coun = resultado8['region'].values[i]
        nomes_paises.append(pycountry.countries.get(alpha_2 = coun).name)
        contagem.append(resultado8['Number_of_movies'].values[i])
    except: 
        continue
df_filmes_paises = pd.DataFrame()
df_filmes_paises['country'] = nomes_paises
df_filmes_paises['Movie_Count'] = contagem
df_filmes_paises = df_filmes_paises.sort_values(by = 'Movie_Count', ascending = False)
df_filmes_paises
```

```python
plt.figure(figsize = (20,8))
sns.barplot(y = df_filmes_paises[:20].country, x = df_filmes_paises[:20].Movie_Count, orient = "h")
for i in range(0,20):
    plt.text(df_filmes_paises.Movie_Count[df_filmes_paises.index[i]]-1,
             i + 0.30,
             round(df_filmes_paises["Movie_Count"][df_filmes_paises.index[i]],2))

plt.ylabel('País')             
plt.xlabel('\nNúmero de Filmes')
plt.title('\nNúmero de Filmes Produzidos Por País\n')
```
